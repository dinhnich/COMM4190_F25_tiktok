## A brief reflection on how you changed your prompting strategies to achieve better inter-rater reliability scores

The first change we had to make was to define the concept well. Anti/Pro/Neutral leaves itself to a world of imagination so by using positice toward/negative toward, it makes the prompt more clear on when to code a 1. Furthermore, we added to neutral words such as off-topic, mixed, or unclear because at the first attempt, the LLM wasn't making the correct decisions on what a neutral comment looked like. Neutral means no opinion but it can also mean something unrelated which we had to define. We also had to add a lot more rules than just the concepts itself as not every comment is as simple as "Lebrons the best" or "Michael Jordan is the worst." For example, nicknames, jokes, emojis, all had to be accounted for and we made sure to put some of those examples. Another thing that was extremely difficult was giving the LLM basketball knowledge/context in order to determine if a comment was positive or negative or neutral. For example, Michael won 6 championships and Lebron has won 4. Michael retired and then came back. Lebron has made the finals 8 straight times. Things that only basketball fans knew had to be given to the LLM so it could understand what a comment meant if it said "getting to the finals means nothing if you dont win."
